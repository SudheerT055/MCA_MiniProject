**Lets Start by underStanding what is "Global Terrorism" means?**

Global terrorism is a form of violence or intimidation that is carried out by non-state actors in order to achieve political, religious, or ideological objectives. It is often conducted across international borders and involves the use of tactics such as bombings, assassinations, kidnappings, hijackings, and cyber-attacks. It is often used to target civilians and other non-combatants in order to create fear and destabilize governments.

**Why do people turn towards terrorism?**

People turn towards terrorism for a variety of reasons, including political or religious extremism, a desire for revenge, a desire to gain power or influence, or a perceived need to protect a certain group or cause. In some cases, people may also turn to terrorism as a response to perceived injustices or social exclusion.

>**Fun Facts:**
>
>Terrorism is not a new phenomenon. In fact, the first recorded terrorist attack dates back to 1st century B.C. in which the [Sicarii](https://en.wikipedia.org/wiki/Sicarii#:~:text=The%20Sicarii%20are%20regarded%20as,America%20to%20describe%20a%20hitman.), a Jewish extremist group, assassinated Roman sympathizers in the Judean region.
>
>  The term terrorism was not widely used until the 1970s, when it was popularized by the media in response to acts of violence committed by groups such as the [IRA](https://en.wikipedia.org/wiki/Irish_Republican_Army), the [Baader-Meinhof gang](https://en.wikipedia.org/wiki/Red_Army_Faction), and the [Symbionese Liberation Army](Symbionese_Liberation_Army).


**The criteria for terrorism**

Violent actions are usually categorised according to the perpetrator, the victim, the method, and the purpose. Different definitions emphasise different characteristics, depending on the priorities of the agency involved. 

In our coverage of terrorism, we rely strongly on data from the Global Terrorism Database (GTD), which defines terrorism as acts of violence by non-state actors, perpetrated against civilian populations, intended to cause fear, in order to achieve a political objective. Its definition excludes violence initiated by governments (state terrorism) and open combat between opposing armed forces, even if theyre non-state actors. ## 5.3 Is there more global terrorism today than there was in the 19th century? <a class="anchor" id="5.3"></a>

```python
count=clean_data.Year.value_counts().reset_index(name="Counts").rename(columns={"index":"Year"}).sort_values(by=("Year")).Counts
color_list=[]
for num in count:
    if num<=2500:
        color_list.append("#ffb601")
    elif num<=5000 and num>2500:
        color_list.append("#f48c06")
    elif num<=7500  and num>5000:
        color_list.append("#e95d05")
    elif num<10000 and num>7500:
        color_list.append("#dd2e02")
    elif num>=10000:
        color_list.append("#d00000")
        
fig,axs= plt.subplots(figsize=(10,15))
plt.gcf().set_facecolor('#f5f6f6')

sns.countplot(data=clean_data,y="Year",palette=color_list,linewidth=1)
for i in axs.containers:
            axs.bar_label(i,color="white",fontsize=10,padding=2)
axs.set_facecolor('#111010')
plt.title("Fig 3.1. Total Terrorist Cases Registered Annually",fontsize=14)
#plt.grid()
elements=[Patch(facecolor='#ffb601',
                        label='< 2,500'),
                Patch(facecolor='#f48c06',
                        label='< 5,000'),
                Patch(facecolor='#e95d05',
                        label='< 7,500'),
                Patch(facecolor='#dd2e02',
                        label='< 10,000'),
                Patch(facecolor='#d00000',
                        label='> 10,000')]
axs.legend(handles=elements,title="Counts",fontsize=12)
plt.tight_layout()
```
